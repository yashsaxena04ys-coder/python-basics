```python
from sqlalchemy import create_engine,Column,Integer,String
from sqlalchemy.orm import declarative_base, sessionmaker
```

```python
engine = create_engine('sqlite:///database.db')
```

```python
Base =  declarative_base()
```

```python
#ORM - Object Relational Mapping
```

```python
class User(Base):
    __tablename__ = "users"
    id = Column(Integer, primary_key=True)
    name = Column(String)
    email = Column(String) 
    password = Column(String)
```

```python
Base.metadata.create_all(engine)
```

```python
Session = sessionmaker(bind=engine)
session = Session()
```

```python
new_user = User(name = "Talwiinder", email = "talwiinder@gmail.com", password="funksong")
session.add(new_user)
session.commit()
```

```python
new_user = User(name="Karan Aujla", email = "aujla@gmail.com",password = "52Bars")
session.add(new_user)
session.commit()
```

```python
users = session.query(User).all()
users
```

```python
for u in users:
    print(f'ID : {u.id}, Name : {u.name}, Password : {u.password}')
```

```python
user = session.query(User).filter_by(email="aujla@gmail.com").first()
print(user.name,user.password)
```

```python
user.name = "Aujla my man"
session.commit()
```

```python
class Student(Base):
    __tablename__ = "Student Data"
    rollno = Column(Integer, primary_key=True)
    name = Column(String)
    course = Column(String)
    email = Column(String)
```

```python
Base.metadata.create_all(engine)
```

```python
new_student1 = Student(name = "Yash", email = "student1@gmail.com", course = "Btech(CSE)")
new_student2 = Student(name = "Harshit", email = "student2@gmail.com", course = "Btech(ECE)")
new_student3 = Student(name = "Sachit", email = "student3@gmail.com", course = "LLB")
new_student4 = Student(name = "Prakhar", email = "student4@gmail.com", course = "LLB")

session.add(new_student1)
session.add(new_student2)
session.add(new_student3)
session.add(new_student4)
session.commit()
```

```python
students = session.query(Student).all()
students
```

```python
for s in students:
    print(f'ID : {s.rollno}, Name : {s.name}, Course : {s.course}')
```

```python
llbstudent = session.query(Student).filter_by(course = "LLB").all()
print(llbstudent.name)
```

```python
for llb in  llbstudent:
    print(llb.rollno,llb.name)
```
