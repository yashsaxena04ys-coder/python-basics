```python
import seaborn as sns
```

```python
df = sns.load_dataset('titanic')
df.head()
```

```python
df.shape
```

```python
sns.histplot(df, x='age',bins=10,color='black', kde='true')
```

```python
import matplotlib.pyplot as plt
sns.histplot(df, x='age',bins=10,color='black', kde='true')
plt.xticks(range(0,81,5))
plt.show()
```

```python
sns.countplot(df, x='sex', palette='viridis')
plt.title('Number of passengers by sex')
plt.show()
```

```python
sns.countplot(df, x='pclass', hue='sex')
plt.title('Number of passengers by class')
plt.show()
```

```python
sns.scatterplot(df, x='age',y='fare',hue='sex',palette='magma')
plt.title('age v/s fare')
plt.show()
```

```python
sns.scatterplot(df, x='age',y='fare',hue='alive',size='pclass',palette='magma')
plt.title('age v/s fare')
plt.show()
```
