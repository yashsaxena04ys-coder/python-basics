```python
import plotly.express as px
import seaborn as sns
```

```python
df = sns.load_dataset('car_crashes')
```

```python
df.head()
```

```python
df.shape
```

```python
px.bar(df, x='abbrev',y='total', title='total car accidents by state',
       labels={'abbrev':'state','total':'total accidents'}, template='plotly_dark')
```

```python
!pip install nbformat
```

```python
plot = px.pie(df, values='speeding', names='abbrev',
              title='speeding accidents by state',template='plotly_dark',
              color_discrete_sequence=px.colors.sequential.Blues
              )
plot.update_traces(textposition='inside', textinfo='percent+label')
```

```python
fig = px.scatter(df, x='alcohol',y='speeding',color='abbrev',title='alcohol vs speeding by state',
           labels={'alcohol':'alcohol-related','speeding':'speeding-related',}, size= 'total')
fig.update_layout(title={'x':0.5})
```

```python
px.scatter_3d(df, x='alcohol',y='speeding',z='not_distracted',color='abbrev',title='alcohol vs speeding by state',
           labels={'alcohol':'alcohol-related','speeding':'speeding-related',}, size= 'total')
```

```python
px.line(df, x='abbrev',y='total',title='Accidents by state', markers=True)
```
